# Regression 
- Linear regression is a fundamental supervised machine learning algorithm and statistical method used to model
- The relationship between a dependent variable (target) and one or more independent variables (predictors) by
- fitting a straight line to the observed data.

```python
import pandas as pd
df=pd.DataFrame({"Hours":[1,2,3,4,5,6,7],"Exam Score":[50,55,65,70,75,80,85]})
df
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
y_pred=lr.predict(x)
import pandas as pd
diff=pd.DataFrame({"Actual":y,"predict":y_pred})
print("Coeffecitn m : ",lr.coef_)
print("Intercept c : ",lr.intercept_)
diff
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y,y_pred)
r2 = r2_score(y, y_pred)
print('Mean squared error:', mse)
print('Mean absolute error:', mae)
print('Coefficient of determination (R^2):', r2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,label="Actual",color='blue')
plt.plot(x,y_pred,color='r',label="predicted")
plt.xlabel("Hours Study")
plt.ylabel("Exam Score")
plt.title("Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
df=pd.read_csv('Book1.csv')
df
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1) 

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
diff=pd.DataFrame({"Actualy":y_test,"Preidct":y_pred})
diff
```

```python
print("Coeffecitn m : ",lr.coef_)
print("Intercept c : ",lr.intercept_)
```

```python
from sklearn import metrics
print("MAE: ",metrics.mean_absolute_error(y_test,y_pred))
print("MSE: ",metrics.mean_squared_error(y_test,y_pred))
print("R2 Score: ",metrics.r2_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,label="Actual",color='blue')
plt.plot(x_test,y_pred,color='r',label="predicted")
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.title("Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```

# min 

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
min_mae=1
min_mse=1
min_r2=1
for i in range(1,43) :
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i) 
    from sklearn.linear_model import LinearRegression
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    mae = metrics.mean_absolute_error(y_test,y_pred)
    mse=metrics.mean_squared_error(y_test,y_pred)
    r2=metrics.r2_score(y_test,y_pred)
    if mae<min_mae :
        min_mae=mae
        maei=i
    if mse<min_mse :
        min_mse=mse
        msei=i
    if r2<min_r2:
        min_r2=r2
        r2i=i
print(min_mae)
print(maei)
print("MEA")
print()
print(min_mse)
print(msei)
print("MSE")
print()
print(min_r2)
print(r2i)
print("R2")
```
