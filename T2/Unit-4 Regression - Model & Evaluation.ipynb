# Regression 
- Linear regression is a fundamental supervised machine learning algorithm and statistical method used to model
- The relationship between a dependent variable (target) and one or more independent variables (predictors) by
- fitting a straight line to the observed data.

## Evalution
- r^2 Score
- Mean absolute Score
- Mean Square Score

```python
import pandas as pd 
df=pd.read_csv("test.csv")
x=df.iloc[:,0:1] # Data frame mate 😁😁
y=df.iloc[:,-1] # series mate 😎😎😎
print(type(x))
print(type(y))
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# just se the Change 🧐🧐🧐
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=1) # test_size=0.1
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print(y_pred) # x na testing ma 2 data hata so 2 values aavse 
```

```python
# ek vaar ma to data split karyo to have 2ji var ma no split
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
y_pred=lr.predict(x)
print(y) 
print("Coeffecitn m : ",lr.coef_)
print("Intercept c : ",lr.intercept_)
```

```python
from sklearn import metrics
print("MAE : ",metrics.mean_absolute_error(y,y_pred))
print("MSE : ",metrics.mean_squared_error(y,y_pred))

print("**********************")
print("R^2 : ",metrics.r2_score(y,y_pred))
```
