```python
import pandas as pd
df=pd.read_csv("car data.csv")
df
```

```python
df.info()
```

```python
df.isna().sum()
# data is ok so go ahead
```

```python
df.drop(columns="Car_Name",inplace=True)
print(df.info())
```

```python
df['age']=2026-df['Year']
df.info()
```

```python
df.drop(columns="Year", inplace=True)
df.info()
```

```python
df.columns
```

```python
df=pd.get_dummies(data=df,drop_first=True)
df.columns
```

```python
y=df["Selling_Price"]
x=df.drop(columns="Selling_Price")
print(x.shape)
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1) 

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
diff=pd.DataFrame({"Actualy":y_test,"Preidct":y_pred})

print("Coeffecitn m : ",lr.coef_)
print("Intercept c : ",lr.intercept_)

from sklearn import metrics
print("\nMAE: ",metrics.mean_absolute_error(y_test,y_pred))
print("MSE: ",metrics.mean_squared_error(y_test,y_pred))
print("R2 Score: ",metrics.r2_score(y_test,y_pred))
```

## FuelConsumptionCo2.csv

```python
import pandas as pd
df=pd.read_csv("FuelConsumptionCo2.csv")
df.head()
```

```python
df.info()
```

```python
df.drop(columns=["MAKE","MAKE","VEHICLECLASS",'TRANSMISSION'],inplace=True)
```

```python
df.info()
```

```python
y=df["CO2EMISSIONS"]
x=df[["ENGINESIZE","CYLINDERS","FUELCONSUMPTION_COMB"]]
print(x.shape)
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.2)
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
df=pd.DataFrame({"Actual":y_test,"Predict":y_pred})
df
```

```python
print("Coeffiecet m: ",lr.coef_)
print("Intercept c: ",lr.intercept_)

from sklearn import metrics
print("MEA : ",metrics.mean_absolute_error(y_test,y_pred))
print("MSA : ",metrics.mean_squared_error(y_test,y_pred))
print("R2 : ",metrics.r2_score(y_test,y_pred))
```
