## unidirected graph

```python
import networkx as nx
import matplotlib.pyplot as plt

g=nx.Graph()

g.add_node(1)
g.add_nodes_from([2,3])
g.add_nodes_from(range(4,7))

g.add_edge(1,2)
g.add_edge(1,1)
g.add_edges_from([(2,3),(3,6),(4,6),(5,6)])

nx.draw_networkx(g,node_size=850,node_color='red',width=2,edge_color='blue')
plt.show()
```

## directed graph

```python
import networkx as nx
import matplotlib.pyplot as plt

g=nx.DiGraph()

g.add_node(1)
g.add_nodes_from([2,3])
g.add_nodes_from(range(4,7))

g.add_edge(1,2)
g.add_edge(1,1)
g.add_edges_from([(2,3),(3,6),(4,6),(5,6)])

nx.draw_networkx(g,node_size=1000,node_color='red',width=2,edge_color='blue')
#or nx.draw(g,with_label=True,node_size=1000,node_color='red',width=2,edge_color='blue') 😏😏
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
g = nx.DiGraph()
L = [[0,1,0,1],[0,0,0,1],[1,0,0,0],[1,0,1,0]]
for i in range(0,4):
    for j in range(0,4):
        if L[i][j] == 1:
            g.add_edge(i,j)
nx.draw(g, with_labels=True, node_size=1000, node_color='red', width=2, edge_color='blue')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
g = nx.DiGraph()

L = [[1,0,0,1,0,1,1],[0,1,1,1,1,1,0],[1,1,1,0,1,0,1]]
g.add_nodes_from([0,1,2])

for i in range(0,3):
    for j in range(0,7):
        if L[i][j] == 1:
            g.add_edge(i,j)
            
nx.draw(g, with_labels=True, node_size=1000, node_color='red', width=2, edge_color='blue')
plt.show()
```

```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
pio.renderers.default='notebook'

data={'Dept':["it",'it','hr','hr','fin','fin'],
     "gender":['m','f','m','f','m','m'],
     "Sal":[50000,-50000,45000,48000,70000,72000]}

df=pd.DataFrame(data)
fig=px.bar(df,x='Dept',y='Sal',title='Dept v/s Sal',text_auto=True,color_discrete_sequence=['red','orange'],color='gender')
fig.show()
```

```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
pio.renderers.default='notebook'

data={'Fruit':['Apples','Apples','Bannanas','Bannanas','Oranges','Oranges'],
     'City':['NYC','LA','NYC','LA','NYC','LA'],
     'Sales':[100,120,80,70,90,110],
      'Profit':[20,25,15,10,18,22],
     'Region':["East","East","East","West","West","West"]}

df=pd.DataFrame(data)
fig=px.bar(df,x='Fruit',
           y='Sales',
           title='Fruit sales analysis',
           text_auto="0.2s",
           color_discrete_map={'NYC':"gold","LA":'midnightblue'},
           color='City',
          text="Sales",
          orientation="v",
          hover_data={"Profit",'Region'},
          pattern_shape="City",
          barmode="group")
fig.show()
```

```python
import pandas as pd
import plotly.express as px

data = {
    "Department": ["IT", "HR", "Finance", "Marketing"],
    "Employees": [40, 20, 25, 15]
}

df = pd.DataFrame(data)
fig = px.pie(
    df,
    values="Employees",
    names="Department",
    color_discrete_sequence=["blue", "green", "orange", "red"],
    title="Employee Distribution by Department"
)

fig.update_traces(
    textinfo="percent+label",
    textposition="inside",
    pull=[0,0.1,0,0]
)
fig.show()
```
