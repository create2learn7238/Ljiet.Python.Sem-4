```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# generating 2d 10x10 matrix AT random
data=np.random.randint(low=1,high=100,size=(10,10))
print("The Datato be ploted :\n")
print(data)
#plotting the heatmap
sns.heatmap(data,vmin=10,vmax=70,cmap='inferno',annot=True,linewidths=1,
            linecolor='black',cbar=False,xticklabels=False,yticklabels=False)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
df=pd.read_csv('auto-mpg.csv')
df['horsepower']=pd.to_numeric(df['horsepower'],errors='coerce')
df=df.dropna(subset=['horsepower'])
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='horsepower',y='mpg',hue='origin',
                style='origin',size='weight',palette='Set1')
plt.show()
```
